```python
import numpy as np
import matplotlib.pyplot as plt

import scipy.misc
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
dx = 0.1e-6
dy = 0.1e-6
dz = 1e-9
dt = 10e-15
```

```python
start_t = 0
stop_t = 5e-12

## Simetría X,Y
start_x = -10e-6
stop_x = 10e-6
start_y = -10e-6
stop_y = 10e-6

start_z = 0
stop_z = 500e-9


x = np.linspace(start_x,stop_x,int(round((stop_x-start_x)/dx)))
y = np.linspace(start_y,stop_y,int(round((stop_y-start_y)/dy)))
z = np.linspace(start_z,stop_z,abs(int(round((stop_z-start_z)/dz))))
t = np.linspace(start_t,stop_t,int(round((stop_t-start_t)/dt)))


X,Y,Z = np.meshgrid(x,y,z)
```

```python
ke = 235 # J/(mK s)
k = 134.5 # J /(m^3 k^2)
G = 5.69e17 # J /(m^3 K s)
Cl = 2.42e6  # J /(m^3 K)
R = 0.88 # A.U.
delta = 20e-9 # nm
ball_len = 100e-9 # nm
T0 = 300 # K
tp = 100e-15 # fs
w0 = 12.4e-6 # um

## Distancia de Rayleigh
n = 1 # Coef Refl Aire
wavelength = 1032e-9 # nm
Zr = n*np.pi* w0**2 / wavelength
F = 2 # J /cm**2

x0 = 0 # Posición central del Spot en x
y0 = 0 # Posición central del Spot en y
```

```python
#
w = w0 * (1 + z**2/Zr**2)**0.5
T = (1/tp) * (4*np.log(2)/np.pi)**0.5 * np.exp(-4*np.log(2)*((t-2*tp)/tp)**2)
```

```python
S = (1-R)/ (delta + ball_len) * F * w0**2/w**2 * np.exp(
    -Z/(delta + ball_len) - (((X-x0)**2 + (Y-y0)**2)/w**2))
S = S.astype(np.float32)
```

```python
Q = np.empty((45, S.shape[0],S.shape[1],S.shape[2]),dtype = np.float32)
```

```python
for i in range(45):
    Q[i] = S*T[i]
    if (S*T[i]).all() < 1e-5:
        break
```
