```python
import numpy as np
import matplotlib.pyplot as plt
```

### Size and Step Size

```python
dz = 0.1 ## Paso en Profundidad  nm
start_z = 0
stop_z = 50

dx = 0.1  ## Paso en Superficie um
start_x = 0
stop_x = 50


alpha = 2

dt = (dx**2)/(4*alpha)

gamma = alpha * dt / dx**2

max_iter_time = 750
```

#### Boundary Conditions

```python
u_top = 0.0
u_left = 00.0
u_bottom = 00.0
u_right = 0.0
```

### Initial Conditions

```python
u_initial = 0
```

```python
u = np.full((int(stop_x/dx), int(stop_x/dz)),u_initial)

u[0:5, :] = u_top
u[:,0] = u_left
u[-1,:] = u_bottom
u[:,-1] = u_right


u[100:200,50:350] = 120
```

```python
plt.imshow(u)
```

```python
u_new = np.ones_like(u) * u_initial
for k in range(0,max_iter_time-1):
    A = u[2: , 1:-1]
    B = u[:-2, 1:-1]
    C = u [1:-1,2:]
    D = u[1:-1,:-2]
    E = u[1:-1,1:-1]
    u_new[1:-1,1:-1] = gamma * (A+B+C+D-4*E) + E
    
    u_new[0, :] = u[0, :]
    u_new[:, 0] = u[:,0]
    u_new[-1,:] = u[-1,:]
    u_new[:,-1] = u[:,-1]
    
    if k % 10 == 0:
        plt.imshow(u_new)
        plt.show()
    u = u_new
```
