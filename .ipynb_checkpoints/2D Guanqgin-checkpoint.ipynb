```python
import numpy as np
import matplotlib.pyplot as plt

import scipy.misc
from scipy import ndimage
import matplotlib.pyplot as plt
```

### Parámetros de Mallado

```python
### RECORREMOS UNA PROFUNDIDAD

dx = 8e-9  ## Paso en Profundidad  nm
start_x = 0
stop_x = 600e-9

### PARA UNA SUPERFICIE
dy = 0.1e-6  ## Paso en Superficie um
start_y = -10e-6
stop_y = 10e-6

### EN UN TIEMPO 
dt = 5e-15  ## Time Step fs
start_t = 0
stop_t = 5e-12


x = np.linspace(start_x,stop_x,int(round((stop_x-start_x)/dx)))
y = np.linspace(start_y,stop_y,int(round((stop_y-start_y)/dy)))
t = np.linspace(start_t,stop_t,int(round((stop_t-start_t)/dt)))

X,Y,tt= np.meshgrid(x,y,t)
```

### Parámetros 

```python
# ------------- Red Electrones -----------------
gamma = 134.5 # J / (m^3 K^2)
Ce_aux = gamma # Notación
ke = 235  # J /(m K s)
g = 5.69e17 # J /(m K^3 s)

#  ----------------------------------------------

# ------------- Red Lattice -----------------
Ci = 2.42e6  # J / (m^3 K)
ke = 235  # J /(m K s)
g = 5.69e17 # J /(m K^3 s)

#  -------------------------------------------


# ------------- Parámetros Láser -----------------
R = 0.88 # Reflectancia A.U.
delta = 20e-9 # Optical Penetration (nm)
ball_len = 100e-9 # Ballistic Length (nm)
F = 5*1e4 # Fluence J/cm^2 -> J/m^2 
tp = 100e-15 # fs
#  ------------------------------------------------------

# ------------- Parámetros Varios -----------------
ys = 3e-6 ## Perfil de Apertura del Láser (um)
T0 = 300 # K Temperatura Ambiente (K)
x0 = 0 # Posición central del Spot en x
y0 = 0 # Posición central del Spot en y
```

#### Creamos Valores de la Fuente y  Vectores de Temperatura

```python
T = np.exp(-4*np.log(2)*((tt-2*tp)/tp)**2)

S = np.sqrt(4*np.log(2)/np.pi) * (1-R) / (tp*(delta + ball_len)) * F * np.exp(
    -X/(delta + ball_len) - ((Y-y0)/ys)**2)

Q = S*T

print(f"Shape de Q: {Q.shape}")
### Y X T

Te = np.ones_like(Q[:,:,0])*300
Ti = np.ones_like(Q[:,:,0])*300

Tenew = np.zeros_like(Te)
Tinew = np.zeros_like(Tenew)

s = dt/Ce_aux
```

```python
for n in range(0,len(t)-1):
    for i in range(1,len(y)-1):
        for j in range(1,len(x)-1):
            Tenew[i,j] = Te[i,j] + ke*s/Te[i,j] * (
                (Te[i+1,j]-2*Te[i,j]+Te[i-1,j])/dy/dy + (Te[i,j+1]-2*Te[i,j]+Te[i,j-1])/dx/dx) - g*s/Te[i,j]* (Te[i,j]-Ti[i,j]) + Q[i,j,n]*s/Te[i,j]
            Tinew[i,j]  = Ti[i,j] + dt/Ci * g * (Te[i,j] - Ti[i,j])
    ### CONDICIONES DE CONTORNO
    Tenew[0,:], Tenew[-1,:], Tenew[:,0], Tenew[:,-1] = Tenew[1,:], Tenew[-2,:], Tenew[:,1], Tenew[:,-2]
    Tinew[0,:], Tinew[-1,:], Tinew[:,0], Tinew[:,-1] = Tinew[1,:], Tinew[-2,:], Tinew[:,1], Tinew[:,-2]
    Te = Tenew
    Ti = Tinew
    if n%10 == 0:
        plt.figure(1)
        plt.gca().invert_yaxis()
        Te_aux = np.transpose(Te)
        Ti_aux = np.transpose(Ti)
        X_aux = np.transpose(X,(1,0,2))
        Y_aux = np.transpose(Y,(1,0,2))
        im1 = plt.pcolor(Y_aux[:,:,n]*1e6,X_aux[:,:,n]*1e9,Te_aux,vmin = 300,vmax = 14000,cmap="RdYlBu_r",)
        plt.title(f"Te t={t[n]*1e15:.2f} fs")
        plt.xlabel(f"Superficie $\mu m$")
        plt.ylabel(f"Profundidad $nm$")
        plt.colorbar(im1)
        plt.pause(0.01)
        plt.figure(2)
        plt.gca().invert_yaxis()
        im2 = plt.pcolor(Y_aux[:,:,n]*1e6,X_aux[:,:,n]*1e9,Ti_aux,vmin = 300,vmax = 6500,cmap="RdYlBu_r")
        plt.title(f"Ti t={t[n]*1e15:.2f} fs")
        plt.xlabel(f"Superficie $\mu m$")
        plt.ylabel(f"Profundidad $nm$")
        plt.colorbar(im2)
        plt.pause(0.01)
        
        print("\n\n")
```
