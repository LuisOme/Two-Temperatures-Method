```python
import numpy as np
import matplotlib.pyplot as plt

import scipy.misc
from scipy import ndimage
import matplotlib.pyplot as plt

import pandas as pd
import os 
```

### Parámetros de Mallado

```python
### RECORREMOS UNA PROFUNDIDAD

dz = 0.01e-6 ## Paso en Profundidad  nm
start_z = 0
stop_z = 3e-6

dx = 0.05e-6  ## Paso en Superficie um
start_x = 0e-6
stop_x = 8e-6

### EN UN TIEMPO 
dt = 7.5e-15  ## Time Step fs
start_t = 0
stop_t = 5e-12 ## Finalizamos a ps


x = np.linspace(start_x,stop_x,int(round((stop_x-start_x)/dx)))
z = np.linspace(start_z,stop_z,int(round((stop_z-start_z)/dz)))
t = np.linspace(start_t,stop_t,int(round((stop_t-start_t)/dt)))

X,Z= np.meshgrid(x,z)
```

### Parámetros Generales Red y Láser

```python
# ------------- Red Electrones -----------------
Ce_aux = 134.5 # J / (m^3 K^2)
ke = 235  # J /(m K s)
g = 5.69e17 # J /(m K^3 s)

#  ----------------------------------------------

# ------------- Red Lattice -----------------
Ci = 2.42e6  # J / (m^3 K)
ki = ke/99 # J (m K^3 s) 
#  -------------------------------------------


# ------------- Parámetros Láser -----------------
R = 0.88 # Reflectancia A.U.
delta = 20e-9 # Optical Penetration (nm)
ball_len = 100e-9 # Ballistic Length (nm)
 
tp = 100e-15 # FWHM fs
w0 = 5e-6 # Waist um A 1/e**2
wavelength = 1.032e-6 # Wavelength (um)
#  ------------------------------------------------------

# ------------- Parámetros Varios -----------------
ys = 3e-6 ## Perfil de Apertura del Láser (um)
T0 = 300 # K Temperatura Ambiente (K)
x0 = 0 # Posición central del Spot en x
n = 1 ## Refractive Index
Zr = n*np.pi*w0**2/wavelength
w = w0*(1+Z**2/Zr**2)**1/2


Tc = 5370 # Temperatura Crítica (K)


cte_e = dt/Ce_aux
cte_i = dt/Ci
```

### Valores para primer pulso (Inicialización)

```python
F = np.linspace(6e4,10e4,5) # Fluence J/cm^2 -> J/m^2
N = 100 # NUMERO DE PULSOS
```

```python
#### CALCULOS PARA DIFERENTES FLUENCIAS
for f in F:
    if (os.path.exists(f"Resultados/Multipulse/Spot_5um/Fluence{f*1e-4:.2f}")== False):
        os.mkdir(f"Resultados/Multipulse/Spot_5um/Fluence{f*1e-4:.2f}")
    
    ### Creamos de Nuevo el Mallado Inicial para otra fluencia.
    ## Valor para cambiar el dz si se va a Z muy grande
    resample = 0
    if f != F[0]:
        dz = 0.01e-6 ## Paso en Profundidad  nm
        start_z = 0
        stop_z = 3e-6

        dx = 0.05e-6  ## Paso en Superficie um
        start_x = 0e-6
        stop_x = 8e-6

        ### EN UN TIEMPO 
        dt = 7.5e-15  ## Time Step fs
        start_t = 0
        stop_t = 5e-12 ## Finalizamos a ps


        x = np.linspace(start_x,stop_x,int(round((stop_x-start_x)/dx)))
        z = np.linspace(start_z,stop_z,int(round((stop_z-start_z)/dz)))
        t = np.linspace(start_t,stop_t,int(round((stop_t-start_t)/dt)))

        X,Z= np.meshgrid(x,z)


        Zr = n*np.pi*w0**2/wavelength
        w = w0*(1+Z**2/Zr**2)**1/2
    
    
    
    depths_craters = []
    diameters_craters = []

        ### ZS PARA PRIMER PULSO ES 0
    Zs = np.zeros_like(x)


    S = (1-R)/(delta + ball_len) * f * w0**2/w**2 * np.exp(
        -(Z-Zs)/(delta+ball_len) - (2*(X-x0)**2/w**2))

    ### PARAMETRO AUXILIAR
    s = dt/Ce_aux

    ### TEMPERATURAS INICIALES
    Te = np.ones_like(X[:,:])*300
    Ti = np.ones_like(X[:,:])*300
    Tenew,Tinew = np.zeros_like(Te), np.zeros_like(Ti)
    
#### CALCULOS PARA MULTIPULSOS
    for nn in range(N):

### CALCULOS T(Z,X) para cada tiempo
        for k,time in enumerate(t):
            
            ### LASER Temporal
            T = 1/tp * np.sqrt(4 * np.log(2)/np.pi) * np.exp(
            -4*np.log(2) * ((time - 2*tp)/tp)**2)
            
            Q = S*T
    
            ## Valores Electrónicos
            A = Te[2:,1:-1]
            B = Te[:-2,1:-1]
            C = Te[1:-1,2:]
            D = Te[1:-1,:-2]
            E = Te[1:-1,1:-1]
            ## Valores Lattice
            G = Ti[2:,1:-1]
            H = Ti[:-2,1:-1]
            I = Ti[1:-1,2:]
            J = Ti[1:-1,:-2]
            K = Ti[1:-1,1:-1]
            Q_reduced = Q[1:-1,1:-1]
            
            Tenew[1:-1,1:-1] = E + cte_e/E * (
            ke/dz**2 * (A+B-2*E) + ke/dx**2 * (C+D-2*E) - g*(E-K) + Q_reduced)
    
            Tinew[1:-1,1:-1] = K + cte_i * (
            ki/dz**2 * (G+H-2*K) + ki/dx**2 * (I+J-2*K) + g*(E-K))
            
            
            Tenew[0,:], Tenew[-1,:], Tenew[:,0], Tenew[:,-1] = Tenew[1,:], Tenew[-2,:], Tenew[:,1], Tenew[:,-2]
            Tinew[0,:], Tinew[-1,:], Tinew[:,0], Tinew[:,-1] = Tinew[1,:], Tinew[-2,:], Tinew[:,1], Tinew[:,-2]
            Te = Tenew
            Ti = Tinew

            if (k == 0 or k == int(len(t)-2)//2 or k == len(t)-2):

                fig, ((ax4)) = plt.subplots(1,1,figsize=(7.5,7.5))
                plt.gca().invert_yaxis()
                fig.tight_layout()

                fig.subplots_adjust(hspace=0.5,wspace=0.5)
                fig.suptitle(f'Pulse {nn+1} for t: {time*1e15:.2f} fs at F: {f*1e-4:.2f} J/cm^2',y = 1.05,fontname="Times New Roman", fontweight="bold")


                im4 = ax4.pcolor(x*1e6,z*1e6,Ti[:,:],vmin = 300,vmax = 6500,cmap="RdYlBu_r",)
                # fig.colorbar(im4)
                ax4.set_title("$T_{lattice}$ Perfil X-Z ")
                ax4.set_xlabel("X ($\mu m$)")
                ax4.set_ylabel("Z ($\mu m$)")

                plt.subplots_adjust(bottom=0.1, right=0.8, top=0.9)
                cax = plt.axes([0.84, 0.05, 0.020, 0.85])
                plt.colorbar(im4,cax=cax)
                plt.show()

        ### OBTENEMOS EL NUEVO PERFIL DE ABLACION    
        X_profile = Ti[0,:]
        for u in range(len(x)):
            for v in range(len(z)):
                ### CALCULO DE Zs y de Profundidad del Pulso
                if Ti[v,u] < 0.9*Tc:
                    z_aux = Z[v,u]
                    # print(z_aux)
                    Zs[u] = z_aux
                    if u == 0:
                        depths_craters.append(z_aux)
                    break
            ### CALCULO DE DIAMETRO DLE PULSO
            if (X_profile[u] < 0.9*Tc):
                diameters_craters.append(2*x[u])
                break
        
        if (len(depths_craters) < nn+1):
            depths_craters.append(z[-1])
            print("MALLADO SUPERADO")
            S = (1-R)/(delta + ball_len) * f * w0**2/w**2 * np.exp(
            -(Z-Zs)/(delta+ball_len) - (2*(X-x0)**2/w**2))


            Te = np.ones_like(Ti[:,:])*300
            Ti = np.ones_like(Ti[:,:])*300
            Tenew,Tinew = np.zeros_like(Te), np.zeros_like(Ti)

            np.save(f"Diameters_{N}_Pulses_{f*1e-4}_Fluence",diameters_craters)
            np.save(f"Depth_{N}_Pulses_{f*1e-4}_Fluence",depths_craters)
        
            break
            
            
        print(f"Depth Crater pulse {nn+1}: {depths_craters[nn]*1e6:.2f} um")
        print(f"Diameter Crater pulse {nn+1}: {diameters_craters[nn]*1e6:.2f} um")

  #### AJUSTAMOS EL MALLADO DE NUEVO
        #### PASAMOS DE Z SI NO USAMOS TODO A UNA MÁS PEQUEÑA
        if (Ti[:int(len(z)*0.75),0] < 350).any():
            print("Reducimos Z")
            stop_z = stop_z*0.75
            z = np.linspace(start_z,stop_z,int(round((stop_z-start_z)/dz)))
            X,Z= np.meshgrid(x,z)
            # Zs = Zs[:len(z),:]

        elif (Ti[:int(len(z)*0.65),0]> 0.7*Tc).all() :
            print("Aumentamos Z")
            stop_z = (stop_z/0.65)                
            z = np.linspace(start_z,stop_z,int(round((stop_z-start_z)/dz)))
            X,Z= np.meshgrid(x,z)
            # Zs = np.concatenate((Zs,Zs),axis = 0)

        if (Ti[0,:int(len(x)*0.9)]< 350).any() :
            print("Reducimos en X")
            stop_x = stop_x*0.9
            x = np.linspace(start_x,stop_x,int(round((stop_x-start_x)/dx)))
            X,Z= np.meshgrid(x,z)
            Zs = Zs[:len(x)]
        elif (Ti[0,:int(len(x)*0.65)]> 0.7*Tc).all():
            print("Aumentamos X")
            stop_x = stop_x/0.65
            x = np.linspace(start_x,stop_x,int(round((stop_x-start_x)/dx)))
            X,Z= np.meshgrid(x,z)
            Zs = np.pad(Zs, (0, len(x)-Zs.shape[0]), 'constant')
        
        w = w0*(1+Z**2/Zr**2)**1/2
        ### CALCULAMOS EL NUEVO POWER LASER
        S = (1-R)/(delta + ball_len) * f * w0**2/w**2 * np.exp(
        -(Z-Zs)/(delta+ball_len) - (2*(X-x0)**2/w**2))

        Te = np.ones_like(X[:,:])*300
        Ti = np.ones_like(X[:,:])*300
        Tenew,Tinew = np.zeros_like(Te), np.zeros_like(Ti)
        
        np.save(f"Resultados/Multipulse/Spot_5um/Fluence{f*1e-4:.2f}/Diameters_{N}_Pulses",diameters_craters)
        np.save(f"Resultados/Multipulse/Spot_5um/Fluence{f*1e-4:.2f}/Depth_{N}Pulses",depths_craters)
    
    np.save(f"Resultados/Multipulse/Spot_5um/Fluence{f*1e-4:.2f}/Diameters_{N}_Pulses",diameters_craters)
    np.save(f"Resultados/Multipulse/Spot_5um/Fluence{f*1e-4:.2f}/Depth_{N}Pulses",depths_craters)
```
