```python
import numpy as np
import matplotlib.pyplot as plt

import scipy.misc
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
Te = np.load("Imagenes/Te_Temporal.npy")
Ti = np.load("Imagenes/Ti_Temporal.npy")

t = np.linspace(0,20e-12,int(20e-12/10e-15))
```

```python
last_value = 1000
plt.title(r'Temperature Evolution F = 2${J}/{cm^{2}}$')

plt.plot(t[:-last_value]*1e12,Te[:-last_value],ls = '--',label = '$T_{e^-}$',color="black")
plt.plot(t[:-last_value]*1e12,Ti[:-last_value],ls = 'dotted',label = '$T_{lattice}$',color="dimgrey")
plt.xlabel("Time (ps)")
plt.ylabel("Temperature(K)")
plt.legend()
plt.savefig('TemperaturesEvolution.svg', format='svg', dpi=1200)
```

```python
Te_sym = np.load("Imagenes/Te_Temporal_Simmetry.npy")
Ti_sym = np.load("Imagenes/Ti_Temporal_Simmetry.npy")

t = np.linspace(0,20e-12,int(20e-12/10e-15))
```

```python
last_value = 1000
plt.title(r'Temperature Evolution F = 2${J}/{cm^{2}}$')

plt.plot(t[:-last_value]*1e12,Te_sym[:-last_value],ls = '--',label = '$T_{e^-}$',color="black")
plt.plot(t[:-last_value]*1e12,Ti_sym[:-last_value],ls = 'dotted',label = '$T_{lattice}$',color="dimgrey")
plt.xlabel("Time (ps)")
plt.ylabel("Temperature(K)")
plt.legend()
plt.savefig('TemperaturesEvolutionSimmetry.svg', format='svg', dpi=1200)
```

```python
last_value = 1000

plt.title(r'Temperature Evolution F = 2${J}/{cm^{2}}$')
plt.plot(t[:-last_value]*1e12,Te[:-last_value],ls = '--',label = '$T_{e^-}$ 3D',color="orange")
plt.plot(t[:-last_value]*1e12,Ti[:-last_value],ls = '--',label = '$T_{lattice}$3D',color="orange")
plt.plot(t[:-last_value]*1e12,Te_sym[:-last_value],ls = 'dotted',label = '$T_{e^-}$ sym',color="dimgrey")
plt.plot(t[:-last_value]*1e12,Ti_sym[:-last_value],ls = 'dotted',label = '$T_{lattice}$ sym',color="dimgrey")

plt.xlabel("Time (ps)")
plt.ylabel("Temperature(K)")
plt.legend()
```
