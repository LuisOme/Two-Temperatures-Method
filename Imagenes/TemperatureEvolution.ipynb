```python
import numpy as np
import matplotlib.pyplot as plt

import scipy.misc
from scipy import ndimage
import matplotlib.pyplot as plt
```

# SIMULACIONES 

### 3D Lenta 

```python
Te = np.load("Imagenes/T-Tiempo/Te_Temporal.npy")
Ti = np.load("Imagenes/T-Tiempo/Ti_Temporal.npy")

t = np.linspace(0,20e-12,int(20e-12/10e-15))

last_value = 1000
plt.title(r'Temperature Evolution F = 2${J}/{cm^{2}}$')

plt.plot(t[:-last_value]*1e12,Te[:-last_value],ls = '--',label = '$T_{e^-}$',color="black")
plt.plot(t[:-last_value]*1e12,Ti[:-last_value],ls = 'dotted',label = '$T_{lattice}$',color="dimgrey")
plt.xlabel("Time (ps)")
plt.ylabel("Temperature(K)")
plt.legend()
plt.savefig('Imagenes/TemperaturesEvolution.svg', format='svg', dpi=1200)
```
