```python
import numpy as np
import matplotlib.pyplot as plt
import os
import re
```

```python
#### ORDENACION 1,2,3,4 ... de los FOLDERS

def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    '''
    alist.sort(key=natural_keys) sorts in human order
    http://nedbatchelder.com/blog/200712/human_sorting.html
    (See Toothy's implementation in the comments)
    '''
    return [ atoi(c) for c in re.split(r'(\d+)', text) ]
```

```python
#### CARGAMOS LOS FOLDERS DE TRABAJO

r = os.listdir()
Fluences = []
for element in r:
    if element.startswith("Fluence",):
        Fluences.append(element)
        
#### GUARDADAS LOS PATHS A LAS FLUENCIAS (STRING)
Fluences.sort(key = natural_keys)

colors = plt.cm.rainbow(np.linspace(0,1,len(Fluences)))

#### GUARDADOS LOS VALORES DE FLUENCIA (FLOAT)
F = []
i = 0
for i,f in enumerate(Fluences):
    F.append(re.findall(r'\d+.\d+', f))
    F[i] = float(F[i][0])
F = np.asarray(F)
```

```python
i = 0
diameter_craters_1_pulse = []
for fluence in Fluences:
    Depth_Folder = os.listdir(fluence)[0]
    Diameter_Folder = os.listdir(fluence)[1]
    Diameter = np.load(fluence+"/"+Diameter_Folder)
    
    plt.scatter(range(1,len(Diameter)+1),Diameter*1e6,color = colors[i],label = f"F:{F[i]:.2f} J/cm$^2$",s=5)
    plt.xlabel("Fluencias (J/cm$^2$)")
    plt.ylabel(fr"Diametro de ablación ($\mu m$)");
    plt.title("Diametro de ablación vs Fluencia. ")
    diameter_craters_1_pulse.append(Diameter[0])
    # print(f"Depth Crater with F:{fluence}: {Diameter[0]*1e6:.4f} um")
    i += 1
    plt.legend()
```

```python
i = 0
depth_craters_1_pulse = []
for fluence in Fluences:
    Depth_Folder = os.listdir(fluence)
    Depth = np.load(fluence+"/"+Depth_Folder[0])
    plt.plot(range(1,len(Depth)+1),Depth*1e6,color = colors[i],alpha = 0.8,ls="--")
    plt.scatter(range(1,len(Depth)+1),Depth*1e6,color = colors[i],label = f"F:{F[i]:.2f} J/cm$^2$",s=2.5,alpha = 0.5,)
    plt.xlabel("NUMERO DE PULSOS")
    plt.ylabel(fr"Diametro de ablación ($\mu m$)");
    plt.title("Profundidad de ablación vs Pulsos. ")
    depth_craters_1_pulse.append(Depth[0])
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    # print(f"Depth Crater with F:{fluence}: {Depth[0]*1e6:.4f} um")
    i += 1
    
    
```

```python
i = 0
for fluence in Fluences:
    # print(fluence)
    Depth_Folder = os.listdir(fluence)[0]
    Depth = np.load(fluence+"/"+Depth_Folder)
    plt.plot(range(1,(5+1)),Depth[:5]*1e6,color = colors[i],alpha = 0.3,ls="--")
    plt.scatter(range(1,5+1),Depth[:5]*1e6,color = colors[i],label = f"F:{F[i]:.2f} J/cm$^2$",s=5,alpha = 1)
    plt.xlabel("NUMERO DE PULSOS")
    plt.ylabel(fr"Diametro de ablación ($\mu m$)");
    plt.title("Profundidad de ablación vs 5 Pulsos ")
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    i += 1
    
```

```python
depth_craters_1_pulse = np.asarray(depth_craters_1_pulse)
plt.plot(F,depth_craters_1_pulse*1e9)
plt.scatter(F,depth_craters_1_pulse*1e9,s=5)
plt.xlabel("Fluencia (J/cm$^2$)")
plt.ylabel("Depth Crater (nm)")
plt.title(" Depth vs Fluencia (1 Pulso)")
```

```python
diameter_craters_1_pulse = np.asarray(diameter_craters_1_pulse)
plt.plot(F,(diameter_craters_1_pulse*1e6)**2,c='orange')
plt.scatter(F,(diameter_craters_1_pulse*1e6)**2,s=5,c='orange')
plt.xscale('log')
plt.xlabel("Fluencia (J/cm$^2$)")
plt.ylabel("Diameter $^2$ Crater ($\mu m$$^2$)")
plt.title(" Diameter$^2$ vs Fluencia (1 Pulso)")
```
