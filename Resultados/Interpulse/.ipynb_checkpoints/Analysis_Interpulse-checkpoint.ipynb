```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Te_evolution = np.load("Te_2.00_Fluence.npy",allow_pickle=True)
Ti_evolution = np.load("Tred_2.00_Fluence.npy")

Te_finish = np.load("Te at 8347.23 ps.npy")
Ti_finish = np.load("Ti at 8347.23 ps.npy")
```

```python
### EN UN TIEMPO 
dt = 2.5e-15  ## Time Step fs
start_t = 0
stop_t = (len(Ti_evolution)-1)*dt ## Finalizamos a um

dz = 0.01e-6 ## Paso en Profundidad  nm
start_z = 0
stop_z = 1e-6

dx = 0.05e-6  ## Paso en Superficie um
start_x = 0
stop_x = 5e-6


x = np.linspace(start_x,stop_x,int(round((stop_x-start_x)/dx)))
z = np.linspace(start_z,stop_z,int(round((stop_z-start_z)/dz)))
t = np.linspace(start_t,stop_t,int(round((stop_t-start_t)/dt)))
```

```python
last_value = 100000
plt.title(r'Temperature Evolution F = 2${J}/{cm^{2}}$')

plt.plot(t[:7500]*1e12,Te_evolution[:7500],ls = '--',label = '$T_{e^-}$',color="blue")
plt.plot(t[:7500]*1e12,Ti_evolution[:7500],ls = '--',label = '$T_{red^-}$',color="black")
# plt.plot(t[0:10000]*1e12,Ti_evolution[:10000],ls = '--',label = '$T_{e^-}$',color="black")
plt.axhline(300,label = "300 K",c='r',ls = "dotted")
# plt.plot(t[:len(Te_evolution)-1]*1e12,Ti_evolution[:-last_value],ls = 'dotted',label = '$T_{lattice}$',color="dimgrey")
plt.xlabel("Time (ps)")
plt.ylabel("Temperature(K)")
plt.legend()
```

```python
last_value = 1
plt.title(r'Temperature Evolution F = 2${J}/{cm^{2}}$')

plt.plot(t[:len(Te_evolution)]*1e9,Te_evolution[:len(Te_evolution)-2],ls = '--',label = '$T_{e^-}$',color="blue")
plt.plot(t[:len(Te_evolution)]*1e9,Ti_evolution[:len(Te_evolution)-2],ls = '--',label = '$T_{red^-}$',color="black")
# plt.plot(t[0:10000]*1e12,Ti_evolution[:10000],ls = '--',label = '$T_{e^-}$',color="black")
plt.axhline(300,label = "300 K",c='r',ls = "dotted")
# plt.plot(t[:len(Te_evolution)-1]*1e12,Ti_evolution[:-last_value],ls = 'dotted',label = '$T_{lattice}$',color="dimgrey")
plt.xlabel("Time (ns)")
plt.ylabel("Temperature(K)")
plt.legend()
```

```python
plt.pcolor(x*1e6,z*1e6,Ti_finish,cmap="RdYlBu_r",vmin = 300)
plt.title(f"Temperatura Lattice para {t[len(Ti_evolution)-2]*1e9:.1f} ns")
plt.xlabel("X ($\mu$m)")
plt.ylabel("Z ($\mu$m)")
plt.colorbar(cmap="RdYlBu_r")
```
