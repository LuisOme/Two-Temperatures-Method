Resolver la ecuaci√≥n:
$$ dy/dt = \frac{(-y + ku)}{\tau} $$ con $$ y_0 = 0$$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def firstorder(y,t,K,u):
    tau = 5.0
    dydt = (-y + K*u) / tau
    return dydt
```

```python
K = 2.0
u = 1.0
t = np.linspace(0,10,20)
y = odeint(firstorder,y0 = 0,t=t,args = (K,u))
```

```python
plt.plot(t,y)
```
