## Q(t,x,y,z)
Lo tenemos definido de esta forma en forma de array también, primer término el tiempo luego 3D [t,x,y,z]

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-30e-6,30e-6,int((40e-6)/0.1e-6))
z = np.linspace(0,100e-6,int((40e-6)/0.01e-6))
Z,X = np.meshgrid(z,x)
```

```python
R = 0.88
delta = 20e-9
ball_len = 100e-9
F = 5


x0 = 0
zs = 0

w0 = 12.4*1e-6
wavelength = 1.032e-6

Zr = np.pi*w0**2/wavelength

w = w0 * (1+Z**2/Zr**2)**0.5
```

```python
S2 = (1-R)/ (delta + ball_len) * w0**2/w**2 * np.exp(-(Z-zs)/(delta+ball_len)-2*(X-x0)**2/w**2)
S2 = S2/S2.max()

S = (1-R)/ (delta + ball_len) * w0**2/w**2 * np.exp(-(Z-zs)/(delta+ball_len)-(X-x0)**2/w**2)
S = S/S.max()
```

```python
plt.plot(x,S[:,0],label = "S(x,z)",ls='dotted',c="b")
plt.plot(x,S2[:,0],label = "S2(x,z)",c="r",ls='dotted')
plt.axhline(1/np.exp(1),c='b',label = "1/e",ls='-.',)
plt.axhline(1/np.exp(2),c='red',label = "1/e**2",ls='-.')
plt.axvline(w0,c = 'green',label = "w0",ls='--')
plt.legend()
```
