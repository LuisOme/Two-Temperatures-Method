```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Te_evolution = np.load("Imagenes/T-Tiempo/Long Time/Te2.50.npy")
Ti_evolution = np.load("Imagenes/T-Tiempo/Long Time/Ti2.50.npy")
```

```python
plt.plot(Te_evolution)
plt.plot(Ti_evolution)
```

```python
t_stop = len(Te_evolution)*10e-15
```

```python
t_stop-5e-12
```

```python
t = np.linspace(5e-12,t_stop,int((t_stop-5e-12)//10e-15))
```

```python
plt.plot(t[:150000],Ti_evolution[:150000],ls="-.",label = "Temperatura Red")
plt.plot(t[:150000],Te_evolution[:150000],ls="--",label = "Temperatura Electrones",alpha=0.5)
plt.axhline(300,c='r',ls="dotted",lw = 0.5,label = "300 K")
plt.xlabel("Tiempo (s)")
plt.ylabel("Temperatura (K)")
plt.legend()
plt.savefig("Imagenes/Interpulse Evolution.png")
```
